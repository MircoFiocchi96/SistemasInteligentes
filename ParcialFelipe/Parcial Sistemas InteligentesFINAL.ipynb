```python
import numpy as np
import seaborn as sns
import pandas as pd
import warnings
from matplotlib import pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage
from scipy.cluster.hierarchy import fcluster
```

```python
def Grubbs(array, umbral):
    if(array.ndim == 1):
        return array[np.abs(Normalize(array))>umbral]

    
def Normalize(array):
    return (array - np.mean(array)) / np.std(array)


def Zscore(array, umbral):
    vec = ((array-np.mean(array))/(np.median(np.abs(array-np.median(array)))))
    return array[np.abs(vec)>umbral]


def Ric(array, k):
    q1 = np.percentile(array,25)
    q3 = np.percentile(array,75)
    ric = q3 - q1
    return array[(array > q3+ric*k) | (array < q1-ric*k)]
```

```python
data = pd.read_excel('BD2.xlsx')
```

```python
print('Cantidad de datos: ', len(data.index), '\n')
print("Cantidad de valores nulos")
print(data.isnull().sum(), '\n')
print("Procentaje de valores nulos respecto a la cantidad total de datos")
print((data.isnull().sum())/len(data.index))
```

```python
print("Debido a que la cantidad de valores nulos es muy baja respecto a la cantidad total de datos, tomaremos la decisión de eliminarlos. Por lo cual se eliminarán 3 filas.", '\n')
data = data.dropna(how='any')
print('Cantidad de datos: ', len(data.index))
```

```python
print("Todos los datos son de tipo estructurado, ya que se encuentran con un formato definido y una organización dentro del archivo .xlsx. Los atributos Upre, U120 y Upos son de tipo numérico discreto ya que se trata de numeros enteros y que no existen valores intermedios entre ellos. Los atributos PP, Uf y Ueq son de tipo numéricos continuos ya que pertenecen a la recta de números reales. Y por último, el atributo Centro es de tipo nominal, ya que determina categorías, en este caso, el lugar físico donde fueron realizados los estudios.")
```

```python
print("Para la utilización de redes neuronales, todos los valores con los que trabajemos deben ser numéricos. El atributo Centro, no es numérico sino que tiene cadena de caracteres, por lo cual será necesario convertirlos en números: 0-IPS y 1-Clínica Asunción")
```

```python
data.replace({'Centro':'IPS'},0,inplace=True)
data.replace({'Centro':'Clínica Asunción'},1,inplace=True)
```

```python
print("Concentración Upre",'\n')
sns.distplot(data['Upre'])
print('Media: ', data['Upre'].mean())
print('Desviación: ', data['Upre'].std())
warnings.filterwarnings('ignore')
```

```python
print("Concentración U120",'\n')
sns.distplot(data['U120'])
print('Media: ', data['U120'].mean())
print('Desviación: ', data['U120'].std())
warnings.filterwarnings('ignore')
```

```python
print("Concentración Upos",'\n')
sns.distplot(data['Upos'])
print('Media: ', data['Upos'].mean())
print('Desviación: ', data['Upos'].std())
warnings.filterwarnings('ignore')
```

```python
print("A partir del gráfico anterior se observa que existen outliers, para corroborar esto haremos lo siguiente:",'\n')
print("Los outliers detectados son:")
print(Zscore(data['Upos'],4))
sns.boxplot(data['Upos'])
```

```python
print("Debido a que los valores son anormales, en el caso del valor 0.6 porque es una concentración tan baja que no es posible. Para el caso del valor 200, puede ser o que el dato haya sido mal medido o que la hemodiálisis no haya sido bien realizada debido a que posterior al tratamiento se espera que la sangre se filtre y la concentración de Urea disminuya. Por tal motivo procederemos a eliminar dichos datos.", '\n')
data = data.drop(labels=[55,121], axis=0)
print('Cantidad de datos: ', len(data.index), '\n')
```

```python
print("Concentración Upos después de la eliminación de outliers",'\n')
sns.distplot(data['Upos'])
print('Media: ', data['Upos'].mean())
print('Desviación: ', data['Upos'].std())
warnings.filterwarnings('ignore')
```

```python
print("Concentración UF",'\n')
sns.distplot(data['UF'])
print('Media: ', data['UF'].mean())
print('Desviación: ', data['UF'].std())
warnings.filterwarnings('ignore')
```

```python
print("Concentración Ueq",'\n')
sns.distplot(data['Ueq'])
print('Media: ', data['Ueq'].mean())
print('Desviación: ', data['Ueq'].std())
warnings.filterwarnings('ignore')
```

```python
print("Concentración PP",'\n')
sns.distplot(data['PP'])
print('Media: ', data['PP'].mean())
print('Desviación: ', data['PP'].std())
warnings.filterwarnings('ignore')
```

```python
print("Intentaremos determinar si la sospecha de que existen dos grupos es correcta, utilizando las concentraciones de los pacientes.")
concentraciones = pd.DataFrame(data.loc[:,['Upre','U120','Upos','Ueq']])
concentraciones.head(5)
```

```python
C = linkage(concentraciones,'ward')

plt.figure(figsize=(25, 10))
plt.title('Dendrograma')
plt.xlabel('Indice de muestra')
plt.ylabel('Distancia')
dendrogram(
    C,
    leaf_rotation=90.,  # rotar el label del eje x
    leaf_font_size=8.,  # tamano de la fuente para el eje x
)
plt.show()
```

```python
print("Del dendrograma anterior se puede deducir que existen dos grupos de pacientes debido ha que existe una distancia amplia entre dichos grupos, muy superior a la distancia presente entre otros grupos.")
```

```python
print("Ahora clasificaremos cada fila de datos de acuerdo a los grupos detectados recientemente.")
k=2  #cantidad de grupos que estamos buscando
groups = fcluster(C, k, criterion='maxclust')
print("Cantidad de datos: ", groups.size,'\n')
print(groups)
```

```python
data['Grupo'] = groups
data.head(5)
```

```python
print("Para caracterizar los perfiles de los grupos podemos observar la siguiente tabla:",'\n')

diffGrupo = data.groupby('Grupo').mean()
print(diffGrupo)
```

```python
print("En la tabla anterior se puede observar que los grupos detectados se caracterizan por la gran diferencia de concentraciones que tienen, el grupo 1 posee concentraciones mucho menores a las del grupo 2.")
```

```python
print("Para determinar si es necesario normalizar los datos, observaremos el siguiente gráfico:",'\n')
con = data.loc[:,['Upre','U120','Upos','Ueq']]
sns.boxplot(data=con)
```

```python
print("En el gráfico anterior, podemos determinar que si bien las concentraciones Upos y Ueq se encuentran en el mismo rango de valores, las concentraciones U120 y Upre se encuentran en rangos de valores más altos. Esta diferencia de rangos hace que cuando calculamos la distancia, las concentraciones como Upre y U120 tengan mayor peso que las otras dos restantes, obteniendo así distancias mayores. Recordemos que la distancia es necesaria para posteriormente realizar el agrupamiento de datos.")
```

```python
print("Para determinar si las características de Peso y Uf difieren sustancialmente entre los dos centros, podemos observar la siguiente tabla:",'\n')
pesoUfCentro = pd.DataFrame(data.loc[:,['PP','UF','Centro']])
diffCentro = pesoUfCentro.groupby('Centro').mean()
print(diffCentro,'\n')
diffc = np.abs(diffCentro.iloc[0] - diffCentro.iloc[1])/len(data.index)
print("La diferencia de PP entre los dos centros es de ",diffc[0]*100,"%, mientras que la diferencia de UF es del ",diffc[1]*100,". Por lo que no existe una diferencia muy grande.")
```

```python
print("Para determinar si las características de Peso y Uf difieren sustancialmente entre los dos grupos, podemos observar la siguiente tabla:",'\n')
pesoUfGrupo = pd.DataFrame(data.loc[:,['PP','UF','Grupo']])
diffGrupo = pesoUfGrupo.groupby('Grupo').mean()
print(diffGrupo,'\n')
diffg = np.abs(diffGrupo.iloc[0] - diffGrupo.iloc[1])/len(data.index)
print("La diferencia de PP entre los dos grupos es de ",diffg[0]*100,"%, mientras que la diferencia de UF es del ",diffg[1]*100,". En cuanto los grupos, podemos decir que la diferencia de PP es mayor que la existente entre los centros, pero sigue sin ser sustancial.")
```
