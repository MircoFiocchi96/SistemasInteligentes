```python
import tensorflow as tf
import keras as ks
import tensorboard as tb
import PIL as P
import sklearn as sklearn
import cv2
import pandas as pd
import numpy as np
import os
from keras.preprocessing.image import ImageDataGenerator
from keras.utils import to_categorical
from keras.callbacks import ModelCheckpoint, TensorBoard
from keras.models import Sequential
from keras.models import Model
from keras.layers import Dense, Dropout, Flatten
from keras.layers import Conv2D, MaxPooling2D
from keras.optimizers import SGD
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from keras.optimizers import Adam
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
import ssl

ssl._create_default_https_context = ssl._create_unverified_context
```

```python
data = pd.read_csv('urea.csv')

data
```

```python
data.isnull().any()
```

```python
d = data[['Upre','U120', 'Upos', 'Ueq']]
d
```

```python
Y = d['Ueq'].values
X = d[['Upre', 'U120', 'Upos']].values

Y = to_categorical(Y)

# labelencoder_y_1 = LabelEncoder()
# Y = labelencoder_y_1.fit_transform(Y)

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3)

X_train, X_val, Y_train, Y_val = train_test_split(X_train, Y_train, test_size=0.28)

print('X train: {} | Y train: {}'.format(X_train.shape, Y_train.shape))
print('X val: {} | Y val: {}'.format(X_val.shape, Y_val.shape))
print('X test: {} | Y test: {}'.format(X_test.shape, Y_test.shape))
      
```

```python
scaler = StandardScaler();
scaler.fit(X_train);

X_train = scaler.transform(X_train)
X_val = scaler.transform(X_val)
X_test = scaler.transform(X_test)
```

```python
input_dim = X_train.shape[1]
nb_classes = 1
exp_name = 'e01'
print('{}  Input_dim {} | classes: {}' .format(exp_name, input_dim, nb_classes))
```

```python
model = Sequential()
model.add(Dense(2, input_dim=input_dim, activation='relu'))
model.add(Dense(1, activation='softmax'))

optimizer = Adam(lr=0.001)
model.compile(loss='sparse_categorical_crossentropy', optimizer=optimizer, metrics=['accuracy'])
model.summary()

experiment_path = os.path.join(' ./experinents/' + exp_name)
if not os.path.exists(experiment_path):
    os.makedirs(experiment_path)
```

```python
callbacks = [
    ModelCheckpoint(os.path.join(experiment_path, 'weights.{epoch:03d}-{val:.2f} hdf5'),
                   monitor='val_acc',
                   save_best_only=False,
                   save_weights_only=False,),
    TensorBoard(log_dir=os.path.join(experiment_path, 'logs'))
]

print('\nTraining...')
h = model.fit(x=X_train, y=Y_train,
             validation_data=(X_val, Y_val),
              epochs=100, 
                batch_size=32,
             callbacks=callbacks)
```

```python
#constantes
rows=224
cols=224
clases=2
test_size=0.2
random_state=2019
epochs=10
bach_size=16

```

```python
mobile = ks.applications.mobilenet_v2.MobileNetV2(input_shape=(224,224,3), 
                                                  include_top=True, 
                                                  weights='imagenet', 
                                                  pooling='avg')
```

```python
mobile.summary()
```

```python
x = mobile.layers[-2].output
predictions = Dense(2, activation='softmax')(x)
model = Model(inputs=mobile.inputs, outputs=predictions)
model.summary()
```

```python
train_datagen = ImageDataGenerator(rescale=1./255,
                                  horizontal_flip=True,
                                  rotation_range=30,
                                  width_shift_range=0.2,
                                  height_shift_range=0.2)

valid_datagen = ImageDataGenerator(rescale=1./255)

train_gen = train_datagen.flow_from_directory('pets',
                                            color_mode='rgb',
                                             target_size=(rows, cols),
                                             batch_size=bach_size)

valid_gen = valid_datagen.flow_from_directory('pets',
                                            color_mode='rgb',
                                             target_size=(rows, cols),
                                             batch_size=bach_size)

```

```python
model.compile(Adam(lr=0.0001), loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
model.fit_generator(train_gen, steps_per_epoch=4, validation_data=valid_gen, 
                    validation_steps=2, epochs=epochs, verbose=2)
```

```python
model.save('cat-dogs.h5')
```

```python
from keras.preprocessing import image
from keras.applications.resnet50 import preprocess_input, decode_predictions
img_path = 'gatito.jpg'
img = image.load_img(img_path, target_size=(224, 224))
x = image.img_to_array(img)
x = np.expand_dims(x, axis=0)
x = preprocess_input(x)

preds = model.predict(x)

img
```
