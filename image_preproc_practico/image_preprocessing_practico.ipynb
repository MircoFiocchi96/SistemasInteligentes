```python
import matplotlib.pyplot as plt
import numpy as np
import cv2
img=cv2.imread('images/bridge_0.05_noisy.jpg')
kernel=np.ones((5,5), np.float32)/25
dst=cv2.dilter2D(img,-1,kernel)
plt.subplot(121),plt.imshow(img),plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(dst),plt.title('Promediada')
plt.xticks([]), plt.yticks([])
plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
img=cv2.imread('images/dl_book.jpg')
```

```python
plt.imshow(img)
```

```python
img= cv2.resize(img, (1900,1900))
```

```python
plt.imshow(img)
```

```python
h, w, _ = img.shape
M = cv2.getRotationMatrix2D((w//2, h//2),195,1)
dst = cv2.warpAffine(img,M,(w,h))
f = plt.figure(figsize=(20,20))
ax= f.add_subplot(121)
ax2=f.add_subplot(122)
ax.imshow(img)
ax2.imshow(dst)
```

```python
h,w,ch= dst.shape
pts1=np.float32([[800,2800], [800,600], [2500,2800],[2500,500]])
pts2=np.float32([[0,300], [0,0], [200,300], [200,0]])

M=cv2.getPerspectiveTransform(pts1,pts2)

dst2=cv2.warpPerspective(dst, M, (200,300))

for x,y in pts1:
    cv2.circle(dst,(x,y),3, (0,255,0), 6)
f= plt.figure(figsize=(20,20))
ax=f.add_subplot(121)
ax2=f.add_subplot(122)
ax.imshow(cv2.cvtColor(dst, cv2.COLOR_BGR2RGB))
ax2.imshow(cv2.cvtColor(dst2, cv2.COLOR_BGR2RGB))

```
